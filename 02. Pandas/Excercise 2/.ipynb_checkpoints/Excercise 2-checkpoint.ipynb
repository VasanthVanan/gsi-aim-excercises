# Data Cleaning, Sort, and Groupby, Join/Combine, Exporting Data

```python
import pandas as pd
import numpy as np
import openpyxl
```

```python
f = pd.read_csv('ipl_player.csv')
```

## Data Cleaning

renaming the column names.

```python
f.isnull()
# check whether the dataframe has null values
# Column '5+ Innings Wickets' has null values
# notnull() - Opposite of f.isnull()
```

```python
f.dropna()
# Drop all rows that contain null values
# as column '5+ Innings Wickets' has null values,
# all the rows are dropped.
```

```python
f.dropna(axis=1)
# Drop all columns that contain null values
# only column '5+ Innings Wickets' is dropped
```

```python
f.fillna('0')
# Replace all null values with '0'
```

```python
f.replace('IPL 2016','IPL 2k16')
# Replace all values equal to first argument with second arguement
```

## Sort, and Groupby

```python
f.sort_values('Highest Score',ascending=False)
# Sort values by col1 in ascending order
```

```python
f.sort_values(['Highest Score','Batting Average'],ascending=[False,False])
# Sort values by col1 in ascending order then col2 in descending order
```

```python
t = f.groupby(['Team','Player'])
# Returns groupby object for values from multiple columns
t.head()
```

```python
score = f.loc[:,['Matches','Catches Taken','Stumpings Made']]
score.head(100)
```

```python
score.apply(np.mean)
# Apply the function np.mean() across each column
```

## Join/Combine

```python
data1 = f.iloc[1:11,[3,27,28]]
data1
# Returns columns as a new DataFrame
```

```python
data2 = f.iloc[110:121,[3,27,28]]
data2
# Returns columns as a new DataFrame
```

```python
data1.append(data2)
# Add the rows in df1 to the end of df2 (columns should be identical)
```

```python
data3 = f.iloc[1:11,:4]
data3
```

```python
data4 = f.iloc[1:11,7:10]
data4
```

```python
result = pd.concat([data3, data4],axis=1)
# Add the columns in df1 to the end of df2 (rows should be identical)
result
```

## Exporting Data

Write to a CSV file

```python
result.to_csv('result.csv')
# to Write to a CSV file
```

```python
result.to_excel('result.xlsx')
# to Write to an Excel file
```

```python
result.to_json('result.json')
# Write to a file in JSON format
```
