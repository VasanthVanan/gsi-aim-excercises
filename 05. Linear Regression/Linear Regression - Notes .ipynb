## Linear Regression:

Regression is a predective modelling technique which is related to a dependent and independent variable. 
It is basically graphing a line over a set of data points

Linear Regression is a method to predict the dependent variable (Y) based on values of independent variable(X). It is in the form of equation of a Straight Line. i.e y = mx + c.

Similarly, Linear regression is expressed in the form of straight line with intercept and slopes.
i.e, Y = (Y intercept) + Slope(X) + error

By using this formula, values are predicted.
Here, Violet represents the actual Value, whereas Brown represents the predicted value.
the distance between the violet and brown are called residual errors

By Determining Least square value, the line is adjusted for the best fit line by minimising the distance bw the line and data point. There are lot of methods to do it.

- sum of squared errors
- sum of absolute errors
- Root mean square error
