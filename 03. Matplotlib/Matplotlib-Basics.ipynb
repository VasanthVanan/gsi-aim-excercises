# MatPlotlib - Data Visualization

It can create bar charts, histogram, scatter-plots and lot more..

```python
from matplotlib import pyplot as plt
import numpy as np
```

## X-Axis Data

```python
x = np.arange(1,11)
```

```python
print(x)
```

## Y-Axis Data

```python
y = np.random.randint(1,50,10)
```

```python
print(y)
```

## Create a Plot

```python
plt.plot(x,y)
```

## Create Linear Graph

```python
x1 = np.arange(1,11)
y1 = np.arange(1,11)
```

```python
plt.plot(x1,y1)
```

## Labelling 

```python
plt.plot(x,y)
plt.title('Graph Title')
plt.xlabel('X Axis Attributes')
plt.ylabel('Y Axis Attributes')
```

## Changing the Line Property

```python
plt.plot(x,y,color='yellow',linewidth=5,linestyle=':')
plt.grid(True)
```

## Compare with two different data

```python
plt.plot(x,y,color='yellow',linewidth=5, linestyle=':')
plt.plot(x1,y1,color='green',linewidth=5)
plt.grid(True)
```

# Bar Graph

create two list with names and values for X-Axis and Y-Axis

```python
fruits = ['Orange','apple','water-melon','mango']
```

```python
rate = [50,75,45,40]
```

```python
plt.bar(fruits,rate, color='brown')
 
# to have horizontal bar, use barh
#plt.barh(fruits,rate, color='brown')

plt.title('Graph Title')
plt.xlabel('X Axis Attributes')
plt.ylabel('Y Axis Attributes')
```

# Scatter Graph

Create lists of data with respective to X and Y

```python
x = list(np.random.randint(1,50,10))
```

```python
y = list(np.random.randint(1,50,10))
```

```python
z = list(np.random.randint(1,50,10))
```

```python
plt.scatter(x,y,color='red')
plt.scatter(x,z,color='green')
```

# Histogram

creates a graph with distribution of continuous variable

```python
import pandas as pd
```

```python
f = pd.read_csv('netflix_titles.csv')
```

### Release Year - Histogram Representation

```python
plt.hist(f['release_year'])
#plt.hist(f['listed_in'])
plt.xlabel('Year')
plt.ylabel('Number of Movies')
```

## Boxplot Representation

```python
f.boxplot(column="release_year")
```
